```python
class Net(nn.Module):
  def _init_(self, input_sz=1):
    super(Net,self)._init_()
    self.layer=nn.Linear(input_sz, 1)
    self.input_sz= input_sz

    def forward(self, x):
      x=self.layer(x)
      return x
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create a figure and axis for the diagram
fig, ax = plt.subplots(figsize=(12, 8))

# Set the diagram title
ax.set_title('Level 1 Diagram of Order Management System', fontsize=16)

# Define positions of processes and external entities
processes_positions = {
    "Receive Order": (0.5, 0.8),
    "Check Inventory": (0.75, 0.6),
    "Process Payment": (0.25, 0.6),
    "Update Inventory": (0.75, 0.4),
    "Generate Invoice": (0.25, 0.4),
    "Ship Order": (0.5, 0.2)
}

entities_positions = {
    "Client": (0.05, 0.8),
    "Inventory": (0.95, 0.6),
    "Supplier": (0.05, 0.4),
    "Employee": (0.05, 0.2),
    "Management": (0.95, 0.2)
}

# Draw rectangles for processes
for process, (x, y) in processes_positions.items():
    ax.add_patch(mpatches.FancyBboxPatch((x-0.1, y-0.05), 0.2, 0.1, boxstyle="round,pad=0.1", fc="lightyellow", ec="black"))
    ax.text(x, y, process, ha='center', va='center', fontsize=12)

# Draw rectangles for external entities
for entity, (x, y) in entities_positions.items():
    ax.add_patch(mpatches.FancyBboxPatch((x-0.1, y-0.05), 0.2, 0.1, boxstyle="round,pad=0.1", fc="lightblue", ec="black"))
    ax.text(x, y, entity, ha='center', va='center', fontsize=12)

# Define arrows to draw between entities and processes
arrows = [
    ("Client", "Receive Order"),
    ("Receive Order", "Check Inventory"),
    ("Receive Order", "Process Payment"),
    ("Check Inventory", "Update Inventory"),
    ("Process Payment", "Generate Invoice"),
    ("Generate Invoice", "Ship Order"),
    ("Update Inventory", "Ship Order"),
    ("Inventory", "Check Inventory"),
    ("Supplier", "Update Inventory"),
    ("Employee", "Process Payment"),
    ("Management", "Generate Invoice")
]

# Function to draw arrows between boxes
def draw_arrow(start, end, ax):
    ax.annotate('', xy=processes_positions[end] if end in processes_positions else entities_positions[end],
                xytext=entities_positions[start] if start in entities_positions else processes_positions[start],
                arrowprops=dict(facecolor='black', arrowstyle='->'))

# Draw arrows between the boxes
for start, end in arrows:
    draw_arrow(start, end, ax)

# Remove axes for cleaner look
ax.axis('off')

# Show the diagram
plt.tight_layout()
plt.show()
```

```python
# Redraw the figure for Level 2 Diagram of "Employee Management" Process from Level 1

# Create a figure and axis for the Level 2 diagram
fig, ax = plt.subplots(figsize=(12, 8))

# Set the diagram title
ax.set_title('Level 2 Diagram of Employee Management Process', fontsize=16)

# Define positions of subprocesses in Employee Management
subprocesses_positions = {
    "Employee Registration": (0.5, 0.8),
    "Attendance Tracking": (0.75, 0.6),
    "Task Assignment": (0.25, 0.6),
    "Performance Review": (0.25, 0.4),
    "Payroll Processing": (0.75, 0.4)
}

entities_positions = {
    "Employee": (0.05, 0.8),
    "HR Dept": (0.05, 0.4),
    "Finance Dept": (0.95, 0.4)
}

# Draw rectangles for subprocesses in Employee Management
for subprocess, (x, y) in subprocesses_positions.items():
    ax.add_patch(mpatches.FancyBboxPatch((x-0.1, y-0.05), 0.2, 0.1, boxstyle="round,pad=0.1", fc="lightyellow", ec="black"))
    ax.text(x, y, subprocess, ha='center', va='center', fontsize=12)

# Draw rectangles for external entities
for entity, (x, y) in entities_positions.items():
    ax.add_patch(mpatches.FancyBboxPatch((x-0.1, y-0.05), 0.2, 0.1, boxstyle="round,pad=0.1", fc="lightblue", ec="black"))
    ax.text(x, y, entity, ha='center', va='center', fontsize=12)

# Define arrows to draw between subprocesses and external entities
arrows_level2 = [
    ("Employee", "Employee Registration"),
    ("Employee Registration", "Attendance Tracking"),
    ("Attendance Tracking", "Task Assignment"),
    ("Task Assignment", "Performance Review"),
    ("Performance Review", "Payroll Processing"),
    ("Payroll Processing", "Finance Dept"),
    ("Performance Review", "HR Dept")
]

# Function to draw arrows between boxes
def draw_arrow_level2(start, end, ax):
    ax.annotate('', xy=subprocesses_positions[end] if end in subprocesses_positions else entities_positions[end],
                xytext=entities_positions[start] if start in entities_positions else subprocesses_positions[start],
                arrowprops=dict(facecolor='black', arrowstyle='->'))

# Draw arrows for subprocess interactions and entity interaction
for start, end in arrows_level2:
    draw_arrow_level2(start, end, ax)

# Remove axes for cleaner look
ax.axis('off')

# Show the diagram
plt.tight_layout()
plt.show()
```
